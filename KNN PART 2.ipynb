```python
data=pd.read_csv('F:/PythonML/Python Datasets/Classification Datasets/Social_Network_Ads.csv')
```

```python
# x as input these 2 columns converted into numpy array store in capital x
X=data.iloc[:,2:4].values
```

```python
#400 row and 2 columns
X.shape
```

```python
#use output variable is fron last columens
Y=data.iloc[:,-1].values
```

```python
Y.shape
```

```python
# divide the data into training and test for cross validation for model slection
#create 4 variables from model slection
from sklearn.model_selection import train_test_split #split data ito training and testing the data
```

```python
# only 20% test data remaining data for training data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20)
```

```python
X.shape
```

```python
# this data for training data
X_train.shape 
```

```python
#this data for test data
X_test.shape
```

```python
#find out distance between 2 point fromm 2 columns very large diff bet age and salary thats why we use 
#class sklearn liabry craete object standardscaler class
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
#this is my age columns and estimated salry columns
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.fit_transform(X_test)
X_test
```

```python
# 1st method is squareroot of 
np.sqrt(X_train.shape[0])
```

```python
# assume value of k is 17
k=17
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=k)
```

```python
# train our model
knn.fit(X_train,Y_train)
```

```python
y_predict1=knn.predict(X_test)
```

```python
y_predict1.shape
```

```python
Y_test.shape
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_predict1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,y_predict1)
```

```python
#using 2nd method 
```

```python
accuracy=[]
for i in range (1,26):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    accuracy.append(accuracy_score(Y_test,knn.predict(X_test)))
```

```python
len(accuracy)
```

```python
plt.plot(range(1,26),accuracy)
```

```python
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,Y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
accuracy_score(Y_test,y_pred)
```
